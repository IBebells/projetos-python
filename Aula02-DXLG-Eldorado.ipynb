Item A

```python
#Receber dados de: nome, data nascimento, sexo, cidade, estado
from datetime import datetime as dt

def recebe_nome ():
  return str(input('Digite seu nome: ')).lower()

def recebe_data ():
  data_nascimento = input('Digite a data do seu nascimento: ').split('/')
  data_nascimento.reverse()
  iso_data_nascimento = '-'.join(data_nascimento)
  data_nascimento = dt.fromisoformat(iso_data_nascimento)
  return data_nascimento.strftime('%y-%m-%d')

def recebe_sexo ():
  return str(input('Digite seu sexo: ')).lower()

def recebe_cidade():
  return str(input('Digite a cidade onde nasceu: ')).lower()

def recebe_estado():
  return str(input('Digite a Estado onde nasceu: ')).lower()
```

```python
def printa_dados ():
  recebe_nome()
  recebe_data()
  print('Seu nome é {nome} do sexo {sexo} nasceu na dia {data} na cidade de {cidade} no Estado {estado}'.format(nome = recebe_nome(),
                                                                                                                 sexo = recebe_sexo(),
                                                                                                                 data = recebe_data(),
                                                                                                                 cidade = recebe_cidade(),
                                                                                                                 estado = recebe_estado()))
```

```python
printa_dados()
```

```python
from random import randrange

def pergunta():
  resposta = str(input('Você quer tentar adivinhar um número S ou N: \n')).upper()
  return resposta

def sorteia_numero():
  return randrange(6)

def recebe_tentativa():
  return int(input('Qual o seu palpite de 0 a 6: '))

def verifica_numero():
  tentativas = 0
  n = sorteia_numero()
  while tentativas < 3:
    tentativa = recebe_tentativa()
    if tentativa == n:
      print('Parabéns! Você acertou!')
      tentativas = 4
    elif tentativas == 2:
      print('Game Over')
      tentativas = 4
    else:
      print('Que pena tente novamente!')
      tentativas += 1
```

```python
if pergunta() == 'S':
  sorteia_numero()
  verifica_numero()
else:
  print('Que pena!')
```

```python
numeros = []
for i in range(3):
  numero = sorteia_numero()
  if numero not in numeros:
    numeros.append(numero)
  else:
    continue

n1, n2, n3 = numeros
print('Os numeros sorteados foram: {n1}, {n2}, {n3}'.format(n1 = n1, n2 = n2, n3 = n3))
```
