```python
#modulo calculadora.py
def mult (n1, n2):
   return n1 * n2

def soma (n1, n2):
   return n1 + n2

def div (n1, n2):
   if n2 == 0:
      print("Não existe divisão por zero!")
   else:
      return n1/n2

def subt (n1, n2):
   return n1-n2

def pot(n1, n2):
   return n1 ** n2

def converte_binario (numero):
   opcao = int(input("Você quer converter um número para [1]Decimal ou [2]Binário: "))
   if opcao == 1:
      return bin(numero)[2:]
   elif opcao == 2:
      return int(numero, 2)

def raiz (radicando, indice):
   return radicando ** (1/indice)
```

```python
#tratamento de erros
#função para trata dados onde se recebe apenas dados positivos e inteiros
def trata_idade(numero):
  while True:
    try:
      numero = int(numero)
    except:
      numero = input("Número inválido. Digite novamente: ")
    else:
      if numero < 0 or numero > 110:
        numero = input("Número inválido. Digite novamente: ")
      else:
        return numero
        break
```

```python
#Digita um texto para receber qualquer dado
def recebe_dado(variavel):
  texto = "Digite " + variavel + ":"
  dado = input(texto)
  return dado
```

```python
#Recebe os dados que colocou na função
def recebe_infos():
  nome = recebe_dado("nome")
  idade = recebe_dado("idade")
  altura = recebe_dado("altura")
  peso = recebe_dado("peso")
  return nome, idade, altura, peso
```

```python
def cal_imc(peso, altura):
    return cal.div(peso, cal.pot(altura,2))
```

```python
class Pessoa():
  def __init__(self, nome, idade, altura, peso):
    #atributos
    self.nome = nome
    self.idade = trata_idade(idade)
    self.altura = altura
    self.peso = peso

  #métodos
  def get_name(self):
    return self.nome

  def get_idade(self):
    return self.idade

  def get_altura(self):
    return self.altura

  def get_peso(self):
    return self.peso

  def imc(self):

    if cal_imc(self.peso, self.altura) < 18.5: status = "Magreza"
    elif cal_imc(self.peso, self.altura) >= 18.5 and cal_imc(self.altura, self.altura) < 25: status = "Normal"
    elif cal_imc(self.peso, self.altura) >= 25 and cal_imc(self.altura, self.altura) < 30: status = "Sobrepeso"
    elif cal_imc(self.peso, self.altura) >= 30 and cal_imc(self.altura, self.altura) < 40: status = "Obesidade"
    elif cal_imc(self.peso, self.altura) >= 40: status = "Obesidade Grave"

    print ("Seu IMC é: {cal_imc:.2f}\nClassificado como {status}.".format(cal_imc = cal_imc(self.peso, self.altura), status = status))
```

```python
class Funcionario(Pessoa):
  def __init__(self, nome, idade, altura, peso, horas_trabalho):
    super().__init__(nome, idade, altura, peso)
    self.horas_trabalho = horas_trabalho

  def get_pagamento(self):
    taxa = 21.50

    if idade < 21:
      pagmento = self.horas_trabalho * taxa
    else:
      pagamento = self.horas_trabalho * (taxa + 2.50)

    return pagamento

  def get_bonus(self, pagamento, bonus):
    return cal.soma(pagamento, bonus)

  def get_desconto(self, pagamento, desconto):
    return cal.subt(pagamento, desconto)

  def get_pagamento_convert(self, pagamento):
    return converte_pagamento(pagamento)
```

```python
class BusinessUnit():
  def __init__(self, account="", businees_unit="", currency=""):
    self.account = account

  def get_account(self):
    return self.account
```

```python
class SpendingAccount():
  def __init__(self, month="", year="", scenario="", currency="", value=""):
    self.month = month
    self.year = year
    self.scenario = scenario
    self.currency = currency
    self.value = value

  def get_month(self):
    return self.month

  def get_year(self):
    return self.year

  def get_scenario(self):
    return self.scenario

  def get_currency(self):
    return self.currency

  def get_value(self):
    return self.value
```
