Tratamento dos Dados

```python
dados.columns = ['data_resposta', 'nome', 'data_nascimento', 'sexo', 'cidade_nascimento', 'estado_nascimento']
```

```python
dados
```

```python
nomes_tratados = []
ultimos_nomes = []

for aluno in dados['nome']:
  nome_separado = aluno.split(' ')
  nome_tratado= f'{nome_separado[-1]}, {nome_separado[0]}'
  ultimo_nome = nome_separado[-1]
  nomes_tratados.append(nome_tratado)
  ultimos_nomes.append(ultimo_nome)

nomes_tratados
```

```python
ultimos_nomes
```

```python
datas_tratadas = []
for data in dados['data_nascimento']:
  data_tratada = data.strftime("%d/%m/%y")
  datas_tratadas.append(data_tratada)

datas_tratadas
```

```python
dados['nomes_tratados'] = nomes_tratados
dados['ultimos_nomes'] = ultimos_nomes
dados['datas_tratadas'] = datas_tratadas
```
